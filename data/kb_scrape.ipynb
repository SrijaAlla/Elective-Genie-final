```python
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
```

```python
urls = [
    # 'https://catalogs.buffalo.edu/content.php?filter%5B27%5D=CSE&filter%5B29%5D=&filter%5Bkeyword%5D=&filter%5B32%5D=1&filter%5Bcpage%5D=1&cur_cat_oid=2&expand=&navoid=67&search_database=Filter#acalog_template_course_filter',
    # 'https://catalogs.buffalo.edu/content.php?catoid=2&navoid=67&filter%5B27%5D=CSE&filter%5B29%5D=&filter%5Bkeyword%5D=&filter%5B32%5D=1&filter%5Bcpage%5D=2&filter%5Bitem_type%5D=3&filter%5Bonly_active%5D=1&filter%5B3%5D=1#acalog_template_course_filter'
    'https://catalogs.buffalo.edu/content.php?filter%5B27%5D=MGO&filter%5B29%5D=&filter%5Bkeyword%5D=&filter%5B32%5D=1&filter%5Bcpage%5D=1&cur_cat_oid=2&expand=&navoid=67&search_database=Filter#acalog_template_course_filter'
]
```

```python
driver = webdriver.Firefox()
```

```python

for url in urls:
    try:
        driver.get(url)

        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')
        table = soup.find('td', class_='block_content')

        links = table.find_all('a')
        
        for link in links:
            link_text = link.get_text().strip()
            print(link_text)
            if '(opens a new window)' in link_text:
                continue
            xpath = f"//a[contains(text(), '{link_text}')]"
            try:
                element = driver.find_element(By.XPATH, xpath)
                element.click()
            except: 
                print('DID NOT FIND: ', link_text)

            time.sleep(1) 

        # driver.quit()
        print("All links clicked successfully.")

        course_info = driver.find_elements(By.CSS_SELECTOR, 'td.coursepadding')

        with open("output.txt", "a") as f:
            for course in course_info:
                print(course.text)
                f.write(course.text + "\n$$\n")  

    except Exception as e:
        print(f"An error occurred: {e}")


driver.quit()
```
